####5. Turn the 311 data into a database. You can use the pandas read_sql command (example [here)]('http://bit.ly/1EoP81v') or the db package from yHat to do this. 

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:')
df.to_sql('sf311',engine)
```

###### 6. Display the first 5 rows of the DB

```python
pd.read_sql("""SELECT * FROM sf311 LIMIT 5""", engine, "case_id")
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
pd.read_sql("""SELECT * FROM sf311 WHERE neighborhood='Haight Ashbury' AND request_type='Sidewalk_Cleaning' LIMIT 10""", engine, "case_id")
```

####8. Which type of complaint was most common? 

```python
pd.read_sql("""SELECT request_type FROM sf311 GROUP BY request_type ORDER BY COUNT(*) DESC LIMIT 1""", engine)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
pd.read_sql("""SELECT neighborhood,request_type FROM sf311 GROUP BY request_type ORDER BY COUNT(*) DESC LIMIT 1""", engine, "neighborhood")
```
