## Implementing Supervised, ***k***-NN Classification

```python
#Create training and test sets.

all_wine_classes = list(wine_df)
wine_features = wine_features[1:]

wine_target = wine_df['Wine'].values
wine_data = wine_df.values
```

```python
#Split the data into Train and Test (30%) using random_state of 42

X_train, X_test, y_train, y_test = train_test_split(wine_data, wine_target, test_size=0.3, random_state=42)
```

```python
#Implement KNN

n_neighbors = range(1,61,1)
scores = []

for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append((clf.score(X_test, y_test)))
    
k = scores.index(max(scores),1)+1
k_score= round(max(scores),6)
```

```python
#Plot k v. k_score

plt.figure(figsize=(18,6))
_ = plt.plot(n_neighbors, scores, linewidth=3.0)
plt.xlabel('k nearest neighbors', fontsize=16)
plt.ylabel('KNN score', fontsize=16)
plt.annotate(("Maximum score: %s\nk=%s" % (k_score,k)), fontsize=14, xy=(k,k_score), xytext=(30,0.75), arrowprops=dict(facecolor='black', shrink=0.05))
```

## Application of Cross Validation

```python
clf = neighbors.KNeighborsClassifier(27, weights='uniform')
scores = cross_val_score(clf, wine_data, wine_target, cv=5)

print("\nScores: %s\n"%(scores))      
print("Accuracy: %0.6f (+/- %0.6f)\n" % (scores.mean(), scores.std() * 2))
```

### Clustering with Un-Scaled Data

```python
# compute distance matrix for unscaled data

X=wine_df.iloc[1:,:].values

distx = squareform(pdist(X, metric='euclidean'))
```

```python
# perform clustering using single-linkage and plot the dendrogram for unscaled data

R = dendrogram(linkage(distx, method='single'), color_threshold=10)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram (Single-Linkage, Unscaled)', fontweight='bold', fontsize=14);
```

```python
# perform clustering using Ward's Method and plot the dendrogram for unscaled data

R = dendrogram(linkage(distx, method='ward'), color_threshold=10)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram (Ward\'s Method, Unscaled)', fontweight='bold', fontsize=14);
```

***Conclusions:***

***Clustering using Ward's method appears to identify similar clusters as in the KMeans results, while single-linkage clustering clearly performs poorly, comparatively. Let's try again with ***scaled*** data.***

### Clustering with ***scaled*** data

```python
# Compute distance matrix for scaled data

distx = squareform(pdist(scale(X), metric='euclidean'))
```

```python
# Cluster using single-linkage

R = dendrogram(linkage(distx, method='single'),color_threshold=10)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram (Single-Linkage, Scaled)', fontweight='bold', fontsize=14);
```

```python
# Using Ward's Method

R = dendrogram(linkage(distx, method='ward'), color_threshold=10)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram (Ward\'s Method, Scaled)', fontweight='bold', fontsize=14);
```

### Scaling the data appears to slightly improve identification of clusters using single-linkage, hierarchical clustering, in contrast to the poor performance of clustering the scaled data using the Ward's method.
